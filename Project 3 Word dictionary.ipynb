## 1. Display the words that begin with a specific letter

```python
def display():
    letter=input('Enter the letter which you want to begin the words')
    for i in cont:
        if i.startswith(letter):
            print(i)
#display()            
```

## 2. Find a specific word it is there in the dictionary or not

```python
def findword():
    word=input('enter the word you want to find').lower() #poWer==> power
    found=False
    for i in cont:
        if i==word:
            found=True
    if found==False:
        print('Word is not there in the dict')
    else:
        print('it is there in the dict')
findword()        
        
        
```

## 3. Display the meaning of  the word

```python
def displaymeaning():
    word=input('Enter the word you want to find the meaning of:').lower()
    found=False
    for key,value in cont.items():
        if key==word:
            found=True
            print(word, 'has the meaning of',value)
    if found==False:
        print('No such word exist')
displaymeaning()
```

## 4.  Display the word meanings for the specific word that is there in the meaning

```python
def displaywordmeaning():
    word=input('Enter the word you want to find the meaning of the word that is there in the meaning').lower()
    found=False
    for key,value in cont.items():
        if key==word or word in value:
            found=True
            print('Word',word , 'Meaning',value)
    if found==False:
        print('No such word exist')
#displaywordmeaning()        
        
    
        
```

##   5. Search and Display the words by number of letters

```python
def displaywordbynumletters():
    num=int(input(" Enter the number to find the meaning of it:"))
    found=False
    for i in cont:
        if len(i)==num:
            print(i)
            found=True
    if found==False:
        print('No word is exist')
#displaywordbynumletters()        
```
