# <span style="color:blue">A lambda function</span> in programming is an anonymous function that is defined using the `lambda` keyword. It allows you to create a small, one-line function without needing to define a formal function using the `def` keyword.


```python
# syntax:lambda arguments : expression
```

#  <span style='color:green'> Function vs lambda function</span>

**Function:**
- Defined using the `def` keyword.
- Can have any number of arguments.
- Can contain multiple expressions and statements.
- Must have a name.
- Accessible anywhere in the code using its name.

**Lambda Function:**
- Defined using the `lambda` keyword.
- Typically used for small, one-line operations.
- Can only have a single expression.
- Does not have a name (anonymous).
- Often used for short-term operations, like within higher-order functions.



```python
def square(x):
    print( x**2)
    print(x + 2)
square(3)
# 3lines of code
```

```python
(lambda x:x**2)(3) # data analysis with use this function more,
```

```python
square(4)
```

#  <span style='color:Gray'> Code examples for lambda function</span>

```python
# -
a=lambda x,y: x-y
a(2,5)
```

```python
# % ,only even numbers
e=lambda x:x%2==0
e(3)
```

```python
# %,only odd numbers
o=lambda x:x%2!=0
o(4)
```

```python
# default paramteres
d=lambda x,y=20:x+y
d(10,40)
```

```python
# keyword argument * ,
k=lambda *x:sum(x)
k(1,2,3,4,5)
```

#  <span style='color:navyblue'> Advantages of lambda functions</span>
`Concise`: Lambda functions allow you to write shorter and more compact code.

`Readability`: For simple operations, lambda functions can make code more readable by keeping the operation close to where it's used.

`No Need for Naming`: You don't need to come up with a name for a lambda function, which can be useful for short-lived operations.


`Functional Programming`: Lambda functions are often used in functional programming paradigms, especially when working with higher-order functions like map, filter, and reduce.
