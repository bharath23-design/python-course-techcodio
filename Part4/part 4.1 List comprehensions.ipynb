# **List comprehensions** 
in Python provide a concise way to create lists by performing an operation on each item in an existing iterable (e.g., a list, tuple, or range) and optionally applying a condition to filter elements. List comprehensions are a powerful and readable way to generate new lists, making your code more compact and expressive.

The syntax for a list comprehension is as follows:

```python
new_list = [expression for item in iterable if condition]
```

- `expression` is the operation or transformation you want to apply to each item in the iterable.
- `item` represents an element in the iterable.
- `iterable` is the source data from which the new list is generated.
- `condition` (optional) is used to filter elements from the iterable based on a specified condition.



````python
# program even
for i in range(10):
    if i %2==0:
        print(i)
        
````

````python
#rule 1 if
[ x for x in range(5) if y%2==0 ]
````

````python
#rule 2 if and else
[ y  if y%2==0 else 'odd' for y in range(5) ]
````

````python
# 1 for 5 times
[ 7 for i in range(5)]
````

````python
ls=[1,2,3,4]
n=[ x if x==1 else 10 for x in ls]
print(n)
````

````python
fruits = ["apple", "banana", "cherry", "kiwi", "mango"]

newlist = [x if x != "banana" else 
           "orange 2" for x in fruits]

print(newlist)
````

````python
fruits = ["apple", "banana", "cherry", "kiwi", "mango"]
newlist = [x.lower() for x in fruits  ] #append for var sequnce

print(newlist)
````
