# What are sets
Set items are unordered, unchangeable(immutable), and do not allow duplicate values.

- created with { }


# creating the sets

```python
s={1,2,3,1,2,3}
s
```

```python
# set constructor
s1=set(['a','b','c'])
s1
```

```python
type(s1)
```

# looping the sets

```python
s1
```

```python
for i in s1:
    print(i)
```

# accessing items

```python
for i in s1:  
    if 'a' ==i:
        print('yes ',i)
```
