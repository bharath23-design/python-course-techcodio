# 1. Single inheritance 
- one child class and one parent class


```python
class parent: 
    def display_parentname(self,name):
        print('this is parent class ',name)
        
class child(parent): #inheriting the parent class
    
    def display_childname(self,childname,name):
        super().display_parentname(name)
        print('This is the child class',childname,name)
c=child()        
c.display_childname('B','A')
```

# 2. Multiple inheritance
- it involve more than one parent class


```python
class parent1:
    def display_parent1name(self,parent1name):
        print('This is parent 1 name',parent1name)
class parent2:
    def display_parent2name(self,parent2name):
        print('This is parent 2 name',parent2name)
        
class child(parent1,parent2):
    def display_childname(self,parent1name,parent2name,childname):
        super().display_parent1name(parent1name)
        super().display_parent2name(parent2name)
        print('This is child name',childname)
```

```python
c=child()
c.display_childname('A','B','C')
```

# 3. Multilevel inheritance
- child class act as a parent for another child class


```python
class grandparent: #parent class
    def grandparent_name(self,grandparentname):
        print('this is grandparent name' ,grandparentname)
   
class parent(grandparent): #childclass1 
    def parent(self,grandparentname ,parentname):
        super().grandparent_name(grandparentname)
        print('this is parent name',parentname)
    
class child(parent): #childclass2
    def child(self,grandparentname, parentname,child):
        super().parent(grandparentname,parentname)
        print(' this is child name',child)
```

```python
g=grandparent()
g.grandparent_name('G')
```

```python
p=parent()
p.parent('G','P')
```

```python
c=child()
c.child('g','p','C')
```

```python
class parent: 
    def parent(self):
        print('this is parent class')
        
class childA(parent):
    def childA(self):
        print(' this is child A class')
        
class childB(parent):
    def childB(self):
        print('this is child B class')
        
class childC(parent):
    def childC(self):
        print('this is child C class')
```

# 5 . Hybrid inheritance
- combination of Heirarchical inheritance and multilevel inheritance


```python
class parent: #parent
    def parent(self):
        print('this is parent class')
        
class childA(parent,childB): #child 1
    def childA(self):
        print(' this is child A class')
        
class childB(parent): #child 2
    def childB(self):
        print('this is child B class')
        
class childC(childA ,childB): # childa and childb
    def childC(self):
        print('this is child C class')
```

```python
c3=childC()
c3.childC()
c3.childA()
c3.childB()
c3.parent()
```
