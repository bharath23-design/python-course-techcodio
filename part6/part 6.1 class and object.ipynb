## Instance:
An object is an instance of a class. When you create an object, you create a specific instance of that class, with its own unique attributes and behaviors.

## Attributes: 
These are the properties or data associated with an object. They are defined within the class and are accessed using dot notation. For instance, in a Person class, name and age could be attributes.
