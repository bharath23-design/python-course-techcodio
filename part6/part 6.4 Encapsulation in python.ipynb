# What is encapsulation
In simpler terms, encapsulation in Python is about grouping data and the methods that work on that data together, and restricting access to the internal details of the class. 

# Advantages

This helps in organizing code, enhancing code readability, and promoting the concept of information hiding.

Data Hiding: 

Security

### Public Member: Accessible anywhere from outside of class.
### Private Member: Accessible within the class
### Protected Member: Accessible within the class and its sub-classes

# Real World Example
#  ATM:

Data (encapsulated): Account balances, transaction history, security measures.

Public methods: Withdraw cash, deposit cash, check balance, transfer funds.

Benefits:
- Users interact with a secure interface, protecting sensitive financial data.
- Prevents unauthorized access to accounts or tampering with transactions.
- Maintains integrity of financial data
