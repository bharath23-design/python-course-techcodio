# what is iterable ?
An iterable is any Python object capable of returning its members one at a time. Iterables include sequences like lists, tuples, and strings, as well as objects like dictionaries and sets

```python
lst={'a':'apple','b':'ball'}
for i in lst.items():
    print(i)
```

# What is iterator
An iterator is an object representing a stream of data. It is an object with a __next__ method that returns the next item in the stream when called, and raises a StopIteration exception when no more data is available. An iterator is also an iterable, but it can be exhausted.

```python
#my_lst=[1,2,3,4,5] => as a stream data
# next()
```

```python
my_list=[1,2,3,4,5,6,7]
my_list
```

```python
#iter
my_iterator=iter(my_list)
my_iterator
```

```python
while True:
    try:
       item=next(my_iterator) 
       print(item)
    except   StopIteration :
        print('No data is available in the my_list')
        break
```

# use of iterator
Efficient Sequential Access:

Iterators provide a way to access elements of a collection one at a time without needing to access the entire collection at once.
This is especially useful for large datasets where loading everything into memory at once would be impractical.

# what is generator
A generator is a special type of iterator created using a function with yield statements. Generators allow you to declare a function that behaves like an iterator. Generators are useful when you need to produce a sequence of values lazily (one at a time and on demand) rather than storing all of them in memory.

The `yield` statement is used to produce a value and pause the function, maintaining its state for the next iteration.


```python
def my_generator():
    my_list = [1, 2, 3, 4, 5]
    for item in my_list:
        yield item

gen = my_generator()
next(gen)
```

```python
next(gen)
```

```python
while True:
    try:
       item=next(gen) 
       print(item)
    except   StopIteration :
        print('No data is available in the my_list')
        break
```

# use of generator
Simplifies the creation of the iterator.
Provides memory efficiency by yielding items one at a time.
Enables lazy evaluation, generating items only when needed.
Maintains state automatically between iterations.
