# What is decorator
Decorators are a very powerful and useful tool in Python since it allows programmers to modify the behaviour of a function or class. Decorators allow us to wrap another function in order to extend the behaviour of the wrapped function, without permanently modifying it. 

# Let's Understand the concept by a story .

Once upon a time in the city of India, there was a magical bakery owned by Bharath. This bakery was famous for its enchanted cookies, which had the power to grant wishes. But Bharath had a secret: he used special magic spells, known as decorators, to add these enchantments to his baked goods.


# The Basics: What is a Decorator?
A decorator in Python is like Bharath’s magic spells. It’s a special function that can modify the behavior of other functions or methods. Think of it as a way to add a little extra magic to an already wonderful recipe.

# Chapter 1: The Plain Cookie
One day, a young helper named joe joined the bakery. he was excited to learn how to bake and use magic. Bharath decided to start with an easy cookie recipe.

```python
def make_cookie():
    print('Making the plain cookie') # plain cake
make_cookie()    
```

# Chapter 2: The Magic Spell (Decorator)
Bharath showed joe a magic spell called add_magic that could be used to enhance the cookie with a wish-granting power.

```python
# add_magic => decorator for that function
def add_magic(func):
    def wrapper():
        print('Sprinkling the magic dust') # adding extra magic to the plain cake
        func()
        print('The cookie is now enchanted!') # added extra magic
    return wrapper 
```

This add_magic function is a decorator. It takes another function (func) and returns a new function (wrapper) that adds some extra steps before and after the original function is called

# Chapter 3:Applying the Spell
To use the magic spell, joe had to apply the decorator to the make_cookie function.

```python
@add_magic
def make_cookie():
    print('Making the plain cookie')
make_cookie()    
```
