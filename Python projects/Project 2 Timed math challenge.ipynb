# Implementing time module in the project

```python
import random
import time
operators=['+','-','*']
min_number=2
max_number=10
total_problems=10

def expression_problem():
    left=random.randint(min_number,max_number)
    right=random.randint(min_number,max_number)
    operator=random.choice(operators)
    expr=str(left) + ' '+ operator + ' ' + str(right)
    answer=eval(expr)
    
    return expr,answer


wrong=0
input('press enter to start')
print('------------------------')

start_time=time.time()

for i in range(total_problems):
    expr,answer=expression_problem()
    while True:
        guess_ans=input('problem #'+str(i + 1)+':'+expr+'=')
        if guess_ans == str(answer):
            break
        wrong +=1    
        
end_time=time.time()
total_time=round(start_time-end_time,2)       
print('Good work !you have finished in',total_time,+' ' +  'seconds!')        
        
        
```
