# <span style="color: #ff9900;">Building the Project</span>


```python
import random
import getpass  # Import the getpass module

class BankAccount:
    def __init__(self, account_number, name, age, password, balance=0):
        self.account_number = account_number
        self.name = name
        self.age = age
        self.password = password
        self.balance = balance

    def deposit(self, amount):
        if amount > 0:
            self.balance += amount
            print(f"Deposited ${amount}. New balance: ${self.balance}")
        else:
            print("Invalid deposit amount.")

    def withdraw(self, amount):
        if 0 < amount <= self.balance:
            self.balance -= amount
            print(f"Withdrew ${amount}. New balance: ${self.balance}")
        else:
            print("Invalid withdrawal amount or insufficient funds.")

    def get_balance(self):
        return self.balance

    def account_details_tuple(self):
        return (self.account_number, self.name, self.age, self.password, self.balance)


def create_account(accounts):
    # Generate a random 12-digit account number
    account_number = random.randint(10**11, 10**12 - 1)

    name = input("Enter your name: ")
    age = int(input("Enter your age: "))
    password = getpass.getpass("Set a password for your account: ")  # Use getpass for secure password input
    initial_balance = float(input("Enter initial balance: "))

    new_account = BankAccount(account_number, name, age, password, initial_balance)
    accounts.append(new_account)

    print(f"Account created successfully! Your account number is: {account_number}")


def enter_existing_account(accounts):
    account_number = int(input("Enter account number: "))
    password = getpass.getpass("Enter password: ")  # Use getpass for secure password input

    for account in accounts:
        if account.account_number == account_number and account.password == password:
            return account

    print("Invalid account number or password.")
    return None


def display_all_accounts(accounts):
    print("\nAll Accounts:")
    for account in accounts:
        print(f"Account Number: {account.account_number}, Name: {account.name}, Age: {account.age}, Balance: ${account.get_balance()}")


# Main program
accounts = []

while True:
    print("\n1. Create a new account")
    print("2. Enter an existing account")
    print("3. Display all accounts")
    print("4. Exit")

    choice = input("Enter your choice (1/2/3/4): ")

    if choice == '1':
        create_account(accounts)

    elif choice == '2':
        existing_account = enter_existing_account(accounts)

        if existing_account:
            while True:
                print("\n1. Deposit money")
                print("2. Withdraw money")
                print("3. Delete account")
                print("4. Exit existing account")

                option = input("Enter your option (1/2/3/4): ")

                if option == '1':
                    amount = float(input("Enter the deposit amount: "))
                    existing_account.deposit(amount)

                elif option == '2':
                    amount = float(input("Enter the withdrawal amount: "))
                    existing_account.withdraw(amount)

                elif option == '3':
                    accounts.remove(existing_account)
                    print("Account deleted successfully!")
                    break

                elif option == '4':
                    break

                else:
                    print("Invalid option. Try again.")

    elif choice == '3':
        display_all_accounts(accounts)

    elif choice == '4':
        print("Exiting program. Goodbye!")
        break

    else:
        print("Invalid choice. Try again.")
```
