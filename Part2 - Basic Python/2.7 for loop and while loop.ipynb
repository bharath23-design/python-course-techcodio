# Python For Loops
A for loop is used for iterating over a sequence (that is either a list, a tuple, a dictionary, a set, or a string).

This is less like the for keyword in other programming languages, and works more like an iterator method as found in other object-orientated programming languages.

With the for loop we can execute a set of statements, once for each item in a list, tuple, set etc.

```python
print('Hello world ')
print('Hello world ')
print('Hello world ')
print('Hello world ')
print('Hello world ')
```

```python
for i in range(5):
    print('Hello world',i)
```

# nested for loops
- A nested loop is a loop inside a loop.

The "inner loop" will be executed one time for each iteration of the "outer loop":

```python
# a loop in another loop
```

```python
for i in range(0,5): # print up 0 to 4
    for j in range(0,3): # print up 0 to 2
        print(i,j) 
```

# for and else

```python
for i in range(6):
    print(i)
else:
    print('Finally done')
```

## Looping Through a String

```python
var='python for loops'
for i in var:
    print(i)
```

```python
var='python for loops'
for i in var:
    if i=='p':
        print('P is encounter in the string')
    elif i=='l':
        print(' l is encounter in the string')
    else:
        print('nothing')
else:
    print('Looping through the string is completed')
```

# The while Loop
With the while loop we can execute a set of statements as long as a condition is true.

```python
i=0 #initialization
while i<6: 
    print(i)
    i+=1
```

```python
i=0 #initialization
while i<6: 
    print(i)
    i+=1
else:
    print('Done running while loop')
```

```python
i=0 #initialization
while i<30:
    i+=3
    if i==9:
        print('it is 9')
    elif i==27:
        print('it is 27')
    else:
        print(i)
else:
    print('done running while loop')
```
