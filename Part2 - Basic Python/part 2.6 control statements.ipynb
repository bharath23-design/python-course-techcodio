```python
age=int(input('enter your age'))
is_indian=False
if age> 18 and is_indian: # and case both statements should be true
    print('Eligible')
else:   
    print('Not Eligible')
```

```python
age=int(input('enter your age'))
is_indian=True
if age> 18 or is_indian: # or case any statements should be true
    print('Eligible')
else:   
    print('Not Eligible')
```

# Nested if else

```python
# a if in another if
a=5
b=12
if a>3:
    print(' a is greater than 3')
    if b<11:
        print(' b is less than 11')
    else:
        print(' b is not less than 11')
else:
    print('a is not greater than 3')
```
