## What are  operators
Operators are used to perform operations on variables and values.

```python
a=4
b=5

a+b

# + is a operator
# a and b are operators
# a+b is known as operation => adddition operation
```

 
## Types of operators

```python
# + is for addition
a=5
b=7
a+b
```

```python
# - is substraction
a-b
```

```python
# * is multiplication
a * b
```

```python
# / for the division
a/b
```

## Identity operators
Identity operators are used to compare the objects, not if they are equal, but if they are actually the same object, with the same memory location:


```python
a=5
b=5
a is b
```

```python
c=10
a is c
```

```python
a is not c
```

```python
a is not b
```


## Membership operators
